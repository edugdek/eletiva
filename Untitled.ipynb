```python
import pandas as pd
```

```python
df = pd.read_csv('gerint_solicitacoes.csv', sep = ";")
```

```python
df.head()
```

```python
from datetime import datetime

dates = df['data_autorizacao'].unique()

def convert_to_date(date_str):
    return datetime.strptime(date_str, '%d/%m/%Y')

dates_sorted = sorted(dates, key=convert_to_date)

print (dates_sorted[1])
print (dates_sorted[-1])
```

```python
horas_na_fila = df['horas_na_fila']
print(horas_na_fila)
```

```python
sexo_horas_na_fila = df[['sexo', 'horas_na_fila']]

total_horas_na_fila_por_sexo = sexo_horas_na_fila.groupby('sexo')['horas_na_fila'].sum()

sexo_mais_horas_na_fila = total_horas_na_fila_por_sexo.idxmax()
total_horas_mais_horas_na_fila = total_horas_na_fila_por_sexo.max()

print(f"Sexo com mais horas na fila: {sexo_mais_horas_na_fila}")
print(f"Total de horas na fila para este sexo: {total_horas_mais_horas_na_fila}")
```

```python
# 5 municipios mais comum
import pandas as pd
import matplotlib.pyplot as plt

municipios_solicitantes_mais_comuns = df['municipio_solicitante'].value_counts().head(5)

plt.figure(figsize=(12, 8))
municipios_solicitantes_mais_comuns.plot(kind='bar')
plt.title('Os 5 Municípios Solicitantes Mais Comuns')
plt.xlabel('Município Solicitante')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sexo e situação mais comum
import pandas as pd
import matplotlib.pyplot as plt

municipios_solicitantes_mais_comuns = df['municipio_solicitante'].value_counts().head(5).index.tolist()

df_filtrado = df[df['municipio_solicitante'].isin(municipios_solicitantes_mais_comuns)]

tabela_contagem = pd.crosstab(df_filtrado['sexo'], df_filtrado['situacao'])

tabela_contagem.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Sexo e Situação Mais Comuns por Município Solicitante')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.xticks(rotation=0)
plt.legend(title='Situação')
plt.show()
```

```python
# Tempo Médio
import pandas as pd
import matplotlib.pyplot as plt

municipios_solicitantes_mais_comuns = df['municipio_solicitante'].value_counts().head(5).index.tolist()

df_filtrado = df[df['municipio_solicitante'].isin(municipios_solicitantes_mais_comuns)]

tempo_medio_espera = df_filtrado.groupby('municipio_solicitante')['horas_na_fila'].mean()

plt.figure(figsize=(12, 6))
tempo_medio_espera.plot(kind='bar')
plt.title('Tempo Médio de Espera (Horas na Fila) por Município Solicitante')
plt.xlabel('Município Solicitante')
plt.ylabel('Tempo Médio de Espera (Horas)')
plt.xticks(rotation=45)
plt.show()
```
